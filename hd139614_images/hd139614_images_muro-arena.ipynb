### Polarimetric differential imaging

```python
npixx, npixy = 1024, 1024

x_mid = (npixx - 1) / 2
y_mid = (npixy - 1) / 2
```

```python
phi_arr = np.zeros((npixx, npixy))

for ix in range(npixx):
    for iy in range(npixy):
        phi_arr[ix,iy] = np.arctan2(ix-x_mid, iy-y_mid)
```

```python
plt.imshow(phi_arr)
```

```python
q_phi_h = q_h * np.cos(2 * phi_arr) + u_h * np.sin(2 * phi_arr)
```

```python
n = 30
plt.imshow(q_phi_h[512-n:512+n,512-n:512+n].T)
plt.colorbar()
```

```python
# r^2 scaling

r2_arr = np.zeros((npixx, npixy))

for ix in range(npixx):
    for iy in range(npixy):
        r2_arr[ix,iy] = (ix - x_mid)**2 + (iy - y_mid)**2
```

```python
plt.imshow(r2_arr)
plt.colorbar()
```

```python
q_phi_h_r2 = q_phi_h * r2_arr
```

```python
n = 60
plt.imshow(q_phi_h_r2[512-n:512+n,512-n:512+n], origin='lower')
plt.colorbar()
```

```python
n = 120
q_phi_h_r2_crop = q_phi_h_r2[512-n:512+n,512-n:512+n]
```

### Residual of 2016 and 2018 images

```python
def qphi(image_q, image_u, scale_r2=True):
    assert image_q.shape == image_u.shape
    npixx, npixy = image_q.shape
    x_mid = (npixx - 1) / 2
    y_mid = (npixy - 1) / 2
    
    phi_arr = np.zeros((npixx, npixy))
    for ix in range(npixx):
        for iy in range(npixy):
            phi_arr[ix,iy] = np.arctan2(ix-x_mid, iy-y_mid)
    
    q_phi = image_q * np.cos(2 * phi_arr) + image_u * np.sin(2 * phi_arr)
    
    if scale_r2:
        r2_arr = np.zeros((npixx, npixy))
        for ix in range(npixx):
            for iy in range(npixy):
                r2_arr[ix,iy] = (ix - x_mid)**2 + (iy - y_mid)**2

        q_phi_r2 = r2_arr * q_phi
        
        return q_phi_r2
    
    else:
        return q_phi
    
def crop(image, halfwidth):
    npixx, npixy = image.shape
    x_mid = int((npixx - 1) / 2)
    y_mid = int((npixy - 1) / 2)
    
    return image[x_mid-halfwidth:x_mid+halfwidth, y_mid-halfwidth:y_mid+halfwidth]
```

```python
q_phi_j_2016 = qphi(q_j, u_j)
q_phi_j_2018 = qphi(q_j_2018, u_j_2018)
```

```python
q_phi_residual = q_phi_j_2018 - q_phi_j_2016
```

```python
plt.imshow(q_phi_residual)
plt.colorbar()
```

```python
q_phi_residual_crop = crop(q_phi_residual, 100)
```

```python
plt.imshow(q_phi_residual, vmax=1e3, vmin=-1e3)
plt.colorbar()
```
